```python
dtype = np.dtype([('filter', 'a2'), ('phase', float), ('mag', float)])
master_file = open('cluster-gloess.dat', 'w')
for star in N3201_use:
    #f_handle = open(N3201_dir+'lcvs/'+star+'.mfit')
    try:
        f_handle = open(lcv_dir+'NGC3201_'+star+'.fit')
    except:
        continue
    star_gloess = np.loadtxt(f_handle, dtype=dtype)
    cluster = np.repeat('N3201', len(star_gloess['mag']))
    star_name = np.repeat(star, len(star_gloess['mag']))
    data_save = np.array(zip(cluster, star_name, star_gloess['filter'], star_gloess['phase'], star_gloess['mag']), 
                        dtype=[('c1', 'a5'), ('c2', 'a10'), ('c3', 'a2'), ('c4', float), ('c5', float)])
    np.savetxt(master_file, data_save, fmt='%s %s %s %7.5f %5.3f')
    f_handle.close()
for star in N7078_use:
    #f_handle = open(N7078_dir+'lcvs/'+star+'.mfit')
    try:
        f_handle = open(lcv_dir+'NGC7078_'+star+'.fit')
    except:
        continue
    star_gloess = np.loadtxt(f_handle, dtype=dtype)
    cluster = np.repeat('N7078', len(star_gloess['mag']))
    star_name = np.repeat(star, len(star_gloess['mag']))
    data_save = np.array(zip(cluster, star_name, star_gloess['filter'], star_gloess['phase'], star_gloess['mag']), 
                        dtype=[('c1', 'a5'), ('c2', 'a10'), ('c3', 'a2'), ('c4', float), ('c5', float)])
    np.savetxt(master_file, data_save, fmt='%s %s %s %7.5f %5.3f')
    f_handle.close()
for star in N6402_use:
    #f_handle = open(N6402_dir+'lcvs/'+star+'.mfit')
    try:
        f_handle = open(lcv_dir+'NGC6402_'+star+'.fit')
    except:
        continue
    star_gloess = np.loadtxt(f_handle, dtype=dtype)
    cluster = np.repeat('N6402', len(star_gloess['mag']))
    star_name = np.repeat(star, len(star_gloess['mag']))
    data_save = np.array(zip(cluster, star_name, star_gloess['filter'], star_gloess['phase'], star_gloess['mag']), 
                        dtype=[('c1', 'a5'), ('c2', 'a10'), ('c3', 'a2'), ('c4', float), ('c5', float)])
    np.savetxt(master_file, data_save, fmt='%s %s %s %7.5f %5.3f')
    f_handle.close()
for star in N6121_use:
    #f_handle = open(N6121_dir+'lcvs/'+star+'.mfit')
    try:
        f_handle = open(lcv_dir+'NGC6121_'+star+'.fit')
    except:
        continue
    star_gloess = np.loadtxt(f_handle, dtype=dtype)
    cluster = np.repeat('N6121', len(star_gloess['mag']))
    star_name = np.repeat(star, len(star_gloess['mag']))
    data_save = np.array(zip(cluster, star_name, star_gloess['filter'], star_gloess['phase'], star_gloess['mag']), 
                        dtype=[('c1', 'a5'), ('c2', 'a10'), ('c3', 'a2'), ('c4', float), ('c5', float)])
    np.savetxt(master_file, data_save, fmt='%s %s %s %7.5f %5.3f')
    f_handle.close()
for star in N5904_use:
    #f_handle = open(N5904_dir+'lcvs/'+star+'.mfit')
    try:
        f_handle = open(lcv_dir+'NGC5904_'+star+'.fit')
    except:
        continue
    star_gloess = np.loadtxt(f_handle, dtype=dtype)
    cluster = np.repeat('N5904', len(star_gloess['mag']))
    star_name = np.repeat(star, len(star_gloess['mag']))
    data_save = np.array(zip(cluster, star_name, star_gloess['filter'], star_gloess['phase'], star_gloess['mag']), 
                        dtype=[('c1', 'a5'), ('c2', 'a10'), ('c3', 'a2'), ('c4', float), ('c5', float)])
    np.savetxt(master_file, data_save, fmt='%s %s %s %7.5f %5.3f')                          
    f_handle.close()
master_file.close()
```
